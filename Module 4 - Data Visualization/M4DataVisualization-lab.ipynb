## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib.pyplot as plt

# Getting the needed data from the database
Query1 = """ Select * from master"""
dfn = pd.read_sql_query(Query1,conn)

# plotting the needed data

plt.hist(dfn["ConvertedComp"])
#plt.ticklabel_format(axis='x', style='plain')
#plt.xlabel("")
#plt.ylabel("")
plt.title("Histogram for ConvertedComp")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

#plt.boxplot(dfn['Age'])
dfn['Age'].plot(kind = 'box')
plt.title('Boxplot of Age')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

dfn.plot(x ='Age' , y = 'WorkWeekHrs', kind = 'scatter',alpha=0.7)
plt.title('Scatter plot of age and workweek hours')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

dfn.plot(kind = 'scatter', x = 'WorkWeekHrs' , y = 'CodeRevHrs', s='Age',alpha=0.5)
plt.title('Bubble plot of workweek hours, codeRevHrs and age')
```

```python
plt.scatter(dfn['WorkWeekHrs'], dfn['CodeRevHrs'], s=dfn['Age'])  # 's' argument specifies the bubble size
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bubble Chart')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df_db = pd.read_sql_query(QUERY,conn)
df_count=df_db['DatabaseDesireNextYear'].value_counts().rename_axis('DatabaseName').reset_index(name='Counts')
df_top5=df_count[:5].copy()
explode = (0.2, 0, 0, 0, 0)
plt.pie(df_top5['Counts'],labels=df_top5['DatabaseName'],autopct='%.2f', explode = explode)
plt.title('Top 5 databases to learn')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Query = """ Select Age, WorkWeekHrs , CodeRevHrs from master where Age between 30 and 35"""
df = pd.read_sql_query(Query,conn)

df_selected = df.groupby('Age').median()

df_selected.plot(kind = 'bar',stacked = True)
plt.title('Median workweek hrs and CodeRevHrs in 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """ SELECT ConvertedComp,Age FROM master where Age between 45 and 60 """
df = pd.read_sql_query(QUERY,conn)
df_sel = df.groupby('Age').median('ConvertedComp')

df_sel.plot(kind = 'line')


plt.title('Median salary in 45-60 age group')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch
FROM master
"""
df_mb = pd.read_sql_query(QUERY,conn)
df_cnt=df_mb['MainBranch'].value_counts().rename_axis('MainBranch').reset_index(name='Counts')

df_cnt.plot(kind='barh')
plt.title('MainBranch')
plt.show()
```
